```python
# This file is to reproduce results in Table 2, Table 3, and Table 4.
# Run the code and enter as prompted.


def analyze(PATH):
    
    total_num = 0
    pred_legit = 0
    pred_malicious = 0

    with open(PATH, encoding='utf-8') as f:
        lines = f.readlines()
        
        for line in lines:
            if 'threshold' in line:
                threshold = int(line.strip('.\n').split(' ')[-1])
                continue
                
            elif 'RSSI' in line:
                total_num += 1
                
                tmp = line.split(' ')
                res_list = [i.strip('\n,:{}\'') for i in tmp]
                
                
                # 1st testbed and 2nd testbed
                try:
                    p5_idx = res_list.index('pixel5')
                    p4a_idx = res_list.index('pixel4a')
                    
                    if (int(res_list[p5_idx+1]) >= threshold) or (int(res_list[p4a_idx+1]) >= threshold):
                        pred_legit += 1
                    else:
                        pred_malicious += 1
                        
                except:
                    pass
                
                
                # 3rd testbed
                try:
                    watch_idx = res_list.index('galaxywatch4(yj6p)')
                    
                    if int(res_list[watch_idx+1]) >= threshold:
                        pred_legit += 1
                    else:
                        pred_malicious += 1
                
                except:
                    pass
            
    return total_num, pred_legit, pred_malicious
            



            
if __name__ == "__main__":
    
    path = './'
    
    while True:
        testbed = int(input('Which testbed? (Enter 1 or 2 or 3) \n  1: the first testbed (two-floor house) \n  2: the second testbed (two-bedroom apartment) \n  3: the third testbed (office) \nUser Input: '))
    
        if testbed not in [1,2,3]:
            print('Invalid Answer! Please enter 1 or 2 or 3. \n')    
        elif testbed == 1:
            print('** The first testbed is selected. **\n\n')
            path = path + 'Table2_firstTestbed/'
            break
        elif testbed == 2:
            print('** The second testbed is selected. **\n\n')
            path = path + 'Table3_secondTestbed/'
            break
        elif testbed == 3:
            print('** The third testbed is selected. **\n\n')
            path = path + 'Table4_thirdTestbed/'
            break
        
    while True:
        speaker = int(input('Which smart speaker? (Enter 1 or 2) \n  1: Amazon Echo Dot\n  2: Google Home Mini \nUser Input: '))
    
        if speaker not in [1,2]:
            print('Invalid Answer! Please enter 1 or 2. \n')
        elif speaker == 1:
            print('** Amazon Echo Dot is selected. **\n\n')
            path = path + 'Echo_'
            break
        elif speaker == 2:
            print('** Google Home Mini is selected. **\n\n')
            path = path + 'GoogleHome_'
            break
        
    while True:
        location = int(input('Which deployment location that the smart speaker is placed? (Enter 1 or 2) \n  1: the first deployment location\n  2: the second deployment location \nUser Input: '))
    
        if location not in [1,2]:
            print('Invalid Answer! Please enter 1 or 2. \n')
        elif location == 1:
            print('** The first deployment location is selected. **\n\n')
            path = path + 'Loc1_'
            break
        elif location == 2:
            print('** The second deployment location is selected. **\n\n')
            path = path + 'Loc2_'
            break

    legitPath = path + 'legitimate.txt'
    malPath = path + 'malicious.txt'
    
    # legitimate as negative, malicious as positive
    # predLegitTrue = true negative, predMaliciousFalse = false positive
    # predLegitFalse = false negative, predMaliciousTrue = true positive
    total_legit, predLegitTrue, predMaliciousFalse = analyze(legitPath)
    total_mal, predLegitFalse, predMaliciousTrue = analyze(malPath)
    
    print('\n\n-------------------------------------- Analysis Results --------------------------------------\n\n')
    print('Total number of legitmate commands:%4d | Correctly predicted:%4d | Mistakenly predicted:%3d' % (total_legit, predLegitTrue, predMaliciousFalse))
    print('Total number of malicious commands:%4d | Correctly predicted:%4d | Mistakenly predicted:%3d' % (total_mal, predMaliciousTrue, predLegitFalse))
    
    
    accuracy = ( predLegitTrue + predMaliciousTrue ) / (total_legit + total_mal)
    precision = predMaliciousTrue / ( predMaliciousTrue + predMaliciousFalse )
    recall = predMaliciousTrue / ( predMaliciousTrue + predLegitFalse )
    
    print('Accuracy: %.2f%% | Precision: %.2f%% | Recall: %.2f%%' % (accuracy*100, precision*100, recall*100))
```
